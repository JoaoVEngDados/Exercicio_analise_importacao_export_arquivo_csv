```python
import pandas as pd
```

```python
#importando dados
df = pd.read_csv(r"C:\Users\joaoe.DESKTOP-53TM428\OneDrive\Documents\Pandas\Pratica_Dados\dados_funcionarios.csv")
print(df.head())
```

```python
df_infor = df.info()
df_infor
```

```python
#2-Analise basica (Média de salário)
df_media =df["salario"].mean().astype(int)
print(df_media)
```

```python
#3-Analise basica (Média por Profissão)
df_media_profissao =df.groupby("profissao")["salario"].mean().astype(int)

print(df_media_profissao)
```

```python
#4-Quantidade de pessoas por cidade
df_quantidade_cidade = df["cidade"].value_counts()
df_quantidade_cidade
```

```python
#5 - Filtros 1 (Pessoas com salario acima de 5000)
df_filtro_1 = df[df["salario"] > 5000][["nome","profissao","salario"]]
df_filtro_1
```

```python
#5- Filtro 2 (Pessoas com mais de 10 anos de experiencia)
df_experiencia = df[df["anos_experiencia"] > 10] [["nome","profissao","anos_experiencia"]] 
print(df_experiencia)
```

```python
#5- Filtro 3( Quantidade de Pessoas por Profissão)

df_pessoas_por_profissao = df.groupby("profissao").size().reset_index(name="quantidade")
print(df_pessoas_por_profissao)
```

```python
#6 - Ordenação 1- Colocando em ordem do maior salario ao menor por ordem alfabetica)
df_odernacao_1 = df.sort_values(by=["nome","salario"],ascending=[True,False])[["nome","salario"]]
df_odernacao_1
```

```python
#6 - Ordenação 2-Colocando em ordem quem tem mais experiencia mostrando o nome e profissao e o tempo de experiencia
df_odernacao_2 = df.sort_values(by="anos_experiencia",ascending=False)[["nome","profissao","anos_experiencia"]]
df_odernacao_2
```

```python
#7 Juntando Tudo Filtrar, Arupar e Ordenar,
# Grupo de pessoa na empresa que recebe  ate 3000 e argupar por profissao em ordem alfabetica

grupo1 = df[df["salario"] <= 3000][["nome","profissao","salario"]].sort_values(by="nome")
grupo1

```

```python
#7 Juntando Tudo Filtrar, Arupar e Ordenar,
# A media de cidade que paga mais em cada profissão
df_cidade_paga_med = df.groupby(["cidade","profissao"]) ["salario"].mean().astype(int).reset_index().sort_values(by=["profissao","salario"],ascending=[True,False]).drop_duplicates(subset="profissao").drop_duplicates(subset="cidade")
print("Media de cidades")
print("---------------------------------")
print(df_cidade_paga_med)
df_cidade_paga_max = df.groupby(["cidade","profissao"]) ["salario"].max().astype(int).reset_index().sort_values(by=["profissao","salario"],ascending=[True,False]).drop_duplicates(subset="profissao").drop_duplicates(subset="cidade")
print("---------------------------------")
print("Cidade que paga mais ")
print("---------------------------------")
print(df_cidade_paga_max)
print("---------------------------------")
print("Cidade que paga menos")
print("---------------------------------")
df_cidade_paga_mix = df.groupby(["cidade","profissao"]) ["salario"].min().astype(int).reset_index().sort_values(by=["profissao","salario"],ascending=[True,False]).drop_duplicates(subset="profissao").drop_duplicates(subset="cidade")
print(df_cidade_paga_mix)
print("---------------------------------")
```
