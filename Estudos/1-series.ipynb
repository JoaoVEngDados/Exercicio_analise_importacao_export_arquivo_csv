```python
#1-Criando uma Serie com dados de vendas mensais
vendas_mensais = pd.Series(
    [2000,2500,2800,3500,4000,4500,5000,5500,6000,7000,5000,3000],
    index=['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Agos','Set','Out','Nov','Dez']
)
```

```python
vendas_mensais
```

```python
#2-Calcular e exibir a media de vendas
media_vendas = vendas_mensais.mean()
print(media_vendas)
```

```python
#3-Indentificando o mes com maior venda
mes_maior_venda = vendas_mensais.idxmax()
mes_maior_venda
```

```python
#4-Exibir as vendas acima da media
vendas_acima_media = vendas_mensais[vendas_mensais > media_vendas]
print(vendas_acima_media)
```
