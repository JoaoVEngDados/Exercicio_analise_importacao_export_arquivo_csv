```python
dados_notas = {
    'Nome' : ['Alice','Bob','David','Eva','Charlie'],
    'Matematica' : [85,90,78,88,92],
    'Fisica' : [80,75,88,85,95],
    'Quimica' : [78,82,85,80,88]
}
```

```python
df_notas = pd.DataFrame(dados_notas)
```

```python
df_notas
```

```python
# Calcular a media de notas
df_notas['Média'] = df_notas[['Matematica','Fisica','Quimica']].mean(axis=1)
df_notas
```

```python
#Indentificar Alunos Aprovados com nota boa(media >= 82.5)
alunos_aprovados = df_notas[df_notas['Média'] >= 82.5]
alunos_aprovados
```

```python
#Exibir a média e os alunos aprovados
df_notas[['Nome','Média']]
```
