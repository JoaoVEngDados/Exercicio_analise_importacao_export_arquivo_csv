```python
import pandas as pd
```

```python
df_produtos = pd.read_csv("dados/ecommerce_user_behavior_8000.csv")
df_produtos
```

```python
df_produtos.dtypes
```

```python
df_produtos['cart_items'] = pd.to_numeric(df_produtos['cart_items'])
df_produtos['discount_seen'] = pd.to_numeric(df_produtos['discount_seen'])
```

```python
#1-Desempenho de vendas por Categoria
vendas_categoria = df_produtos.groupby('device_type')['cart_items'].sum()
vendas_categoria
```

```python
#2-Produto mais vendido
produto_mais_vendido = df_produtos.groupby('device_type')['cart_items'].sum().idxmax()
produto_mais_vendido
```

```python
#3-Preço Médio por Produto
media_preco_produto = df_produtos.groupby('device_type')['time_on_site'].mean()
media_preco_produto
```

```python
#5-correlação entre unidades vendidas e preço Unitário
correlacao = df_produtos['cart_items'].corr(df_produtos['ad_clicked'])
print(correlacao)
```

```python
vendas_categoria_df = vendas_categoria.reset_index()
vendas_categoria_df_columms = ['device_type','cart_items']
```
