```python
import pandas as pd
```

```python
df = pd.read_json(r"C:\Users\joaoe.DESKTOP-53TM428\OneDrive\Documents\Pandas\Estudos\receitas.json",encoding="utf-8")
df
```

```python
#Analise Total de Ingredientes por Receitas
receitas_expandidas = []
for receita in df['receitas']:
   total_ingredientes = len(receita['ingredientes'])
   receita_info = {
      'titulo': receita['titulo'],
      'tempo_preparo' : receita['tempo_preparo'],
      'porcoes' : receita['porcoes'],
      'total_ingredientes' : total_ingredientes
   }
   receitas_expandidas.append(receita_info)
```

```python
print(receitas_expandidas)
```

```python
df_expandido = pd.DataFrame(receitas_expandidas)
df_expandido
```
